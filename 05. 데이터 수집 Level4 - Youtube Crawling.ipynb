```python
# 모듈 가져오기.
from selenium import webdriver as wd
import platform, time, urllib
```

```python
# 2. 브라우저 띠우기.
if platform.system() == 'Darwin': # Mac.
    # 맥 PC는 드라이버 있는 위치에서 아래 명령 터미널에서 수행 후 진행.
    # $ xattr -d com.apple.quarantine chromedriver
    driver = wd.Chrome('c:/chromedriver.exe')
    
else: # Window.    
    driver = wd.Chrome('c:/chromedriver.exe')
```

```python
# 3. Youtube 접속.
keyword     = urllib.parse.quote('로제')
target_site = f'https://www.youtube.com/results?search_query={keyword}&sp=CAISBAgCEAE%253D'

# keyword에 따른 해당 url 접속.
driver.get( target_site )
```

```python
# 4. 더보기 액션을 작동시키 1일치(24시간) 데이터를 전부 로드.
# 상위 버전 => 더보기 api 통신 프로토콜을 분석하여 직접 호출.(조심, 많은 검토가 필요.)
# Scroll bar를 밑으로 내리는 행위, 스크립트에서 Scroll 이동 처리.
# 우선 10회 정도만 내리도록 설정.
for n in range(10) :
    
    # 현재 화면에서 스크립트를 가동.
    # 1000 넉넉하게 크기를 설정함.
    driver.execute_script(''' window.scrollBy(0, 1000) ''')
    
    # 대기 3초.
    # 기계적인 Scroll 내림을 감지할 수 있으므로 Sleep 시간을 랜덤하게 주는 것도 좋음.
    time.sleep(3)
```

```python
# 5. 썸네일, 제목, 링크, 조회수, 업로드 시간, 게시자, 영상 요약, 댓글...등
# 영상의 메타 데이터 수집.
# 검색 결과는 ytd-video-renderer 태그를 모으면 된다
videos = driver.find_elements_by_tag_name('ytd-video-renderer')
len(f'하루 동안 변경 된 수 : {videos}')
```

```python
# 최신 2건만 진행.
for video in videos[:2] :
    # 영상 찾기.
    target = video.find_element_by_id('video-title')
    
    # 제목. : a Tag
    title  = target.get_attribute('title')
    
    # 썸네일.
    thumbnail = video.find_element_by_id('img').get_attribute('src')
    
    # URL 링크.
    video_url = target.get_attribute('href')
    print( title, thumbnail, video_url )
```

```python
# 데이터를 담을 리스트 생성.
Video_metas = []
for video in videos[:2] :    
    target    = video.find_element_by_id('video-title')
    title     = target.get_attribute('title')
    thumbnail = video.find_element_by_id('img').get_attribute('src')    
    video_url = target.get_attribute('href')
    
    # 데이터 수집.
    Video_metas.append( {
        't':title,
        'n':thumbnail,
        'v':video_url
    } )
    
Video_metas
```

```python
# 댓글 로드.
for page in Video_metas[:1] :
    
    # 콘텐츠의 상세 페이지로 이동.
    driver.get( page['v'] )
    time.sleep(5)
    
    # 리뷰를 모두 로드하기 위해서 컨텐츠를 스크롤.
    # 우선 10개의 댓글만 가져옴.
    for n in range(10) :
        driver.execute_script('''window.scrollBy(0, 1000)''')
        time.sleep(3)
        
    # 댓글 수집.
    reviews = driver.find_elements_by_tag_name('ytd-comment-thread-renderer')
    for review in reviews:
        print('-'*100)
        # 해당 요소 밑에 있는 모든 텍스트를 가져옴.
        print( review.text  )
        print('-'*100)
```
