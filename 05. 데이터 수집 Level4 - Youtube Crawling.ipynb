```python
# 5. ì¸ë„¤ì¼, ì œëª©, ë§í¬, ì¡°íšŒìˆ˜, ì—…ë¡œë“œ ì‹œê°„, ê²Œì‹œì, ì˜ìƒ ìš”ì•½, ëŒ“ê¸€...ë“±
# ì˜ìƒì˜ ë©”íƒ€ ë°ì´í„° ìˆ˜ì§‘.
# ê²€ìƒ‰ ê²°ê³¼ëŠ” ytd-video-renderer íƒœê·¸ë¥¼ ëª¨ìœ¼ë©´ ëœë‹¤
videos = driver.find_elements_by_tag_name('ytd-video-renderer')
len(f'í•˜ë£¨ ë™ì•ˆ ë³€ê²½ ëœ ìˆ˜ : {videos}')
```

```python
# ìµœì‹  2ê±´ë§Œ ì§„í–‰.
for video in videos[:2] :
    # ì˜ìƒ ì°¾ê¸°.
    target = video.find_element_by_id('video-title')
    
    # ì œëª©. : a Tag
    title  = target.get_attribute('title')
    
    # ì¸ë„¤ì¼.
    thumbnail = video.find_element_by_id('img').get_attribute('src')
    
    # URL ë§í¬.
    video_url = target.get_attribute('href')
    print( title, thumbnail, video_url )
```

```python
# ë°ì´í„°ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸ ìƒì„±.
Video_metas = []
for video in videos[:2] :    
    target    = video.find_element_by_id('video-title')
    title     = target.get_attribute('title')
    thumbnail = video.find_element_by_id('img').get_attribute('src')    
    video_url = target.get_attribute('href')
    
    # ë°ì´í„° ìˆ˜ì§‘.
    Video_metas.append( {
        't':title,
        'n':thumbnail,
        'v':video_url
    } )
    
Video_metas
```

```python
# ëŒ“ê¸€ ë¡œë“œ.
for page in Video_metas[:1] :
    
    # ì½˜í…ì¸ ì˜ ìƒì„¸ í˜ì´ì§€ë¡œ ì´ë™.
    driver.get( page['v'] )
    time.sleep(5)
    
    # ë¦¬ë·°ë¥¼ ëª¨ë‘ ë¡œë“œí•˜ê¸° ìœ„í•´ì„œ ì»¨í…ì¸ ë¥¼ ìŠ¤í¬ë¡¤.
    # ìš°ì„  10ê°œì˜ ëŒ“ê¸€ë§Œ ê°€ì ¸ì˜´.
    for n in range(10) :
        driver.execute_script('''window.scrollBy(0, 1000)''')
        time.sleep(3)
        
    # ëŒ“ê¸€ ìˆ˜ì§‘.
    reviews = driver.find_elements_by_tag_name('ytd-comment-thread-renderer')
    for review in reviews:
        print('-'*100)
        # í•´ë‹¹ ìš”ì†Œ ë°‘ì— ìˆëŠ” ëª¨ë“  í…ìŠ¤íŠ¸ë¥¼ ê°€ì ¸ì˜´.
        print( review.text  )
        print('-'*100)
```
