```python
import os, re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import platform, time, urllib, random
from selenium import webdriver as wd
```

```python
# 환경 설정.
# Chrome 브라우저 띠우기.
chromedriver ='c:/chromedriver.exe'

# Mac 환경.
if platform.system() == 'Darwin': 
    # 맥 PC는 드라이버 있는 위치에서 아래 명령 터미널에서 수행 후 진행.
    # $ xattr -d com.apple.quarantine chromedriver
    driver = wd.Chrome(chromedriver)
    driver.implicitly_wait(3)

# Window 환경.  
else:   
    driver = wd.Chrome(chromedriver)
    driver.implicitly_wait(3)
```

```python
url = "https://www.bloodinfo.net/bloodstats_stocks.do" 
driver.get(url)
time.sleep(3)

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')

# 년도 클릭.
driver.find_element_by_xpath('//*[@id="sel_year_rbc"]').click()
value = random.uniform(2, 5) 
time.sleep(value)

# 반복 작업이 많지 않아, 코드를 구현하지 않고 임의적으로 바꾸어줌.
# 2020년으로 고정. 2021년은 option[1]로 전환.
driver.find_element_by_xpath('//*[@id="sel_year_rbc"]/option[1]').click()
value = random.uniform(2, 5) 
time.sleep(value)

# 년도 설정 클릭.
driver.find_element_by_xpath('//*[@id="sel_year_rbc"]').click()
value = random.uniform(2, 5) 
time.sleep(value)

# 특이한 구조.
# 1 ~ 12 까지 반복. 1 : 작년 말(12월) 기준으로 수집함.
# 2021년 1을 가져와야 2020년 12월 데이터 수집이 가능.
for i in range(1, 5) :
    # 월 버튼을 클릭.
    driver.find_element_by_xpath('//*[@id="sel_month_rbc"]').click()
    time.sleep(2)

    driver.find_element_by_xpath(f'//*[@id="sel_month_rbc"]/option[{i}]').click()
    time.sleep(2)
    
    driver.find_element_by_xpath('//*[@id="frm_search_rbc"]/div/div/input').click()
    time.sleep(2)
   
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    
    a1 = soup.select('#contents_box > div.content > div > div.area_content > div > div > \
                       div > div.linebox.txtc.mb30 > table > tbody')
    print(a1)
    t_list = []
    for i in a1 :
        a2 = i.text.split()
        t_list.append(a2)
    day    = t_list[0][0 :: 3]
    amount = t_list[0][2 :: 3]

    df = pd.DataFrame({'날짜': day , '양': amount})
    
    if os.path.exists('data.csv') == False :
        df.to_csv('data.csv', encoding='utf-8-sig', index=False)  
    else :
        df.to_csv('data.csv', encoding='utf-8-sig', index=False, header=None, mode='a')
```
